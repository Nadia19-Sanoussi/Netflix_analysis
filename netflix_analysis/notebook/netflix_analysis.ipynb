```python
year_counts = df['release_year'].value_counts().sort_index()
threshold = 10
years_small = year_counts[year_counts < threshold].index
df['year_hybrid'] = df['release_year'].apply(lambda x: 'Other' if x in years_small else x)
df
```

```python
# Count titles per year
year_counts = df['release_year'].value_counts().sort_index()

# Threshold for grouping
threshold = 50
years_small = year_counts[year_counts < threshold].index

# Replace small years by 'Other'
df['year_hybrid'] = df['release_year'].apply(lambda x: 'Other' if x in years_small else x)

# Order for plotting: numbers first, then 'Other'
years_order = sorted([y for y in df['year_hybrid'].unique() if y != 'Other']) + ['Other']

# Plot
plt.figure(figsize=(14,5))
sns.countplot(x='year_hybrid', data=df, order=years_order)
plt.xticks(rotation=90)
plt.title("Nombre de films/sÃ©ries par annÃ©e (small years grouped)")
plt.ylabel("Nombre de titres")
plt.xlabel("AnnÃ©e")
plt.show()
```

 # 1. Distribution de la durÃ©e (films vs sÃ©ries)

```python
# Films (minutes)
movies = df[df['type']=="Movie"].copy()
movies['duration_num'] = pd.to_numeric(movies['duration'].str.extract(r'(\d+)')[0], errors='coerce')

plt.figure(figsize=(10,5))
sns.histplot(movies['duration_num'], bins=40, kde=False, color='skyblue')
plt.title("Distribution de la durÃ©e des films (minutes)")
plt.xlabel("DurÃ©e (minutes)")
plt.ylabel("Nombre de films")
plt.show()
```

conclusion:sur netflix la duree des films est en moyenne de 100mins. pourquoi? est ce que la taille d'un film influence les consomateurs

```python
shows = df[df['type']=="TV Show"].copy()
shows['seasons_num'] = pd.to_numeric(shows['duration'].str.extract(r'(\d+)')[0], errors='coerce')

plt.figure(figsize=(8,4))
shows['seasons_num'].value_counts().sort_index().plot(kind='bar', color='salmon')
plt.title("Distribution du nombre de saisons des sÃ©ries")
plt.xlabel("Nombre de saisons")
plt.ylabel("Nombre de sÃ©ries")
plt.show()
```

 3. Genres par pays (top 5 pays) 

```python

top_countries = df['country'].value_counts().index[:5]
subset = df[df['country'].isin(top_countries)]

plt.figure(figsize=(12,6))
sns.countplot(data=subset, x="country", hue="genre",
              order=top_countries,
              hue_order=subset['genre'].value_counts().index[:6])
plt.title("Genres dominants dans les 5 principaux pays")
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

conclusion:  unitedd states sans grandes surprises est a la tete du classement et produit tout genre surtout ls documentaires qui n'est pas fais beaucoup ailleurs. suivit de l'inde a ma grande surprise le bollywood explose autant que Hollywood.

 4. Tendance dâ€™ajouts par genre

```python

year_genre = df.groupby(['year_added','genre']).size().reset_index(name='count')
top_genres = df['genre'].value_counts().index[:5]
subset = year_genre[year_genre['genre'].isin(top_genres)]

plt.figure(figsize=(12,6))
sns.lineplot(data=subset, x="year_added", y="count", hue="genre", marker="o")
plt.title("Ã‰volution des ajouts par genre (top 5)")
plt.ylabel("Nombre de titres")
plt.show()
```

conclusion: les films et series drama sont les plus populaires sur le marche donc ona produit plus et y'en a beaucoup plus sur netflix. 

5. DÃ©lai entre sortie et ajout sur Netflix

```python

df['delay'] = df['year_added'] - df['release_year']
plt.figure(figsize=(10,5))
sns.histplot(df['delay'].dropna(), bins=40, kde=False, color='purple')
plt.title("DÃ©lai entre sortie et ajout sur Netflix")
plt.xlabel("AnnÃ©es aprÃ¨s sortie")
plt.ylabel("Nombre de titres")
plt.show()
```
